```python
import io
s = (
    "animals,n_legs,entry\n"
    "Flamingo,2,2022-03-01\n"
    "Horse,4,2022-03-02\n"
    "Brittle stars,5,2022-03-03\n"
    "Centipede,100,2022-03-04"
)
print(s)
source = io.BytesIO(s.encode())
```

```python
from pyarrow import csv

csv.read_csv(source)
```
